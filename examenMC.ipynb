1. Importa las librerias necesarias

```python
#Cargar librerias
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

2.Lee el archivo CSV desde la URL proporcionada y guardale en tu DataFrame

```python
df= pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/diamonds.csv')
```

3.Imprime las ultimas 5 filas del dataset y genera un resumen estadistico descriptivo de las columnas numéricas

```python
print(df.tail(5))
print(df.describe())
```

4.Un objeto físico no puede tener dimensiones nulas. Encuentra las filas donde la dimension x,y o z sea exactamente igual a 0 y elimina esas filas por completo del DataFrame


```python
df_non_null=df.query('x>0 and y>0 and z>0')
print(df_non_null)
```

5.Crea una nueva columna llamada volumen_aprox calculando el volumen espacial de cada pieza (multiplicando las columnas x * y * z).


```python
df['volumen_aprox']=df['x']*df['y']*df['z']
print(df)
```

Parte 2. Visualizacion de Datos (Matplotlib) Genera las siguientes gráficas. Todas deben incluir título y etiquetas en los ejes X y Y.

1.Gráfico de Caja (Boxplot): Muestra la distribución estadística de la columna price (precio) para identificar visualmente si existen valores atípicos (outliers).

```python
plt.boxplot(df['price'])
plt.title('Grafico de caja')
plt.xlabel('Precio')
plt.ylabel('Y')
plt.show
```

2.Gráfico de Dispersión (Scatter): Grafica el volumen_aprox en el eje X contra el price en el eje Y.

```python
plt.scatter(df['volumen_aprox'],df['price'])
plt.title('volumen aprox vs precio')
plt.xlabel('Volumen aprox')
plt.ylabel('Precio')
plt.show
```

3.Gráfico de Barras: Muestra la cantidad total de piezas que existen para cada tipo de corte (cut). Pista: Necesitarás contar las ocurrencias de cada categoría antes de graficar.

```python
conteo_cortes = df["cut"].value_counts()

plt.figure()

plt.bar(conteo_cortes.index, conteo_cortes.values)

plt.title("Cantidad de Diamantes por Tipo de Corte")
plt.xlabel("Tipo de Corte")
plt.ylabel("Cantidad de Piezas")

plt.show()
```
